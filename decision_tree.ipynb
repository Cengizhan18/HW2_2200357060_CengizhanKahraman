```python
#all libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
# Plot feature importance
import numpy as np
import seaborn as sns
from itertools import combinations
```

```python
#visualizing the features in groups of two
# Column names
columns = ['feature1', 'feature2', 'feature3', 'feature4', 'label']

# Read the data (if there's no header, use header=None)
data_path = r"C:\Users\kahra\Desktop\data\data1.txt"
df = pd.read_csv(data_path, header=None, names=columns)

# Show the first few rows
print(df.head())

# Get the features
features = columns[:-1]  # Last column is the class/label
label = columns[-1]

# Visualize feature pairs
combs = list(combinations(features, 2))

plt.figure(figsize=(15, 10))
for idx, (feat1, feat2) in enumerate(combs, 1):
    plt.subplot(2, 3, idx)
    sns.scatterplot(data=df, x=feat1, y=feat2, hue=label, palette='Set1')
    plt.title(f"{feat1} vs {feat2}")
plt.tight_layout()
plt.show()
```

```python
# Split features and labels
X = df.drop("label", axis=1)
y = df["label"]

# Optional: scale the features (not mandatory for decision trees, but clean)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into train (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Define hyperparameter combinations to test
criteria = ["gini", "entropy"]
max_depths = [3, 5, 10]
min_samples = [2, 5]

```

```python
# Loop over all combinations and train/evaluate the models
for criterion in criteria:
    for depth in max_depths:
        for min_split in min_samples:
            print(f"\n--- Decision Tree (criterion={criterion}, max_depth={depth}, min_samples_split={min_split}) ---")
            clf = DecisionTreeClassifier(
                criterion=criterion,
                max_depth=depth,
                min_samples_split=min_split,
                random_state=42
            )
            clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)

            # Evaluate the model
            print(classification_report(y_test, y_pred, digits=4))

            # Plot confusion matrix
            disp = ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test, cmap='Blues')
            disp.ax_.set_title(f'Confusion Matrix ({criterion}, depth={depth})')
            plt.tight_layout()
            plt.show()

#Train final model for tree visualization and feature importance
final_clf = DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=42)
final_clf.fit(X_train, y_train)




```

```python
#decision plots for diffrent values and feature importance
# Plot the decision tree
plt.figure(figsize=(16, 8))
plot_tree(final_clf, feature_names=df.columns[:-1], class_names=["Fake", "Authentic"], filled=True)
plt.title("Decision Tree Visualization")
plt.show()

importances = final_clf.feature_importances_
features = df.columns[:-1]
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(8, 5))
plt.bar(range(len(importances)), importances[indices], align="center")
plt.xticks(range(len(importances)), features[indices])
plt.title("Feature Importance")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.tight_layout()
plt.show()


# Final model with criterion='entropy' for tree visualization
final_clf_entropy = DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=42)
final_clf_entropy.fit(X_train, y_train)

# Plot decision tree (entropy)
plt.figure(figsize=(16, 8))
plot_tree(final_clf_entropy, feature_names=df.columns[:-1], class_names=["Fake", "Authentic"], filled=True)
plt.title("Decision Tree (criterion = 'entropy', max_depth = 5)")
plt.show()

# Plot feature importance (entropy)
importances_entropy = final_clf_entropy.feature_importances_
features = df.columns[:-1]
indices_entropy = np.argsort(importances_entropy)[::-1]

plt.figure(figsize=(8, 5))
plt.bar(range(len(importances_entropy)), importances_entropy[indices_entropy], align="center")
plt.xticks(range(len(importances_entropy)), features[indices_entropy])
plt.title("Feature Importance (criterion = 'entropy')")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.tight_layout()
plt.show()


# Final model with criterion='gini' for tree visualization
final_clf_gini = DecisionTreeClassifier(criterion="gini", max_depth=5, random_state=42)
final_clf_gini.fit(X_train, y_train)

# Plot decision tree (gini)
plt.figure(figsize=(16, 8))
plot_tree(final_clf_gini, feature_names=df.columns[:-1], class_names=["Fake", "Authentic"], filled=True)
plt.title("Decision Tree (criterion = 'gini', max_depth = 5)")
plt.show()

# Plot feature importance (gini)
importances_gini = final_clf_gini.feature_importances_
indices_gini = np.argsort(importances_gini)[::-1]

plt.figure(figsize=(8, 5))
plt.bar(range(len(importances_gini)), importances_gini[indices_gini], align="center")
plt.xticks(range(len(importances_gini)), features[indices_gini])
plt.title("Feature Importance (criterion = 'gini')")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.tight_layout()
plt.show()
```
